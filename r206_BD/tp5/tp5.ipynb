```python
import pandas as pandadou
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pandadou.read_csv('./datation.csv')
```

```python
x = df.to_numpy()
```

```python
plt.plot(x, color="red")
plt.xlabel("Années en Miliers")
plt.ylabel("Taux de carbone 14")
plt.title("Données de dataion au carbone 14")

y=np.log(x)
plt.plot(y, color="blue")
plt.xlabel("Années en Miliers")
plt.ylabel("Taux de carbone 14")
plt.title("Données de dataion au carbone 14")
plt.show()
```
