```python
def cov(X, Y):
    X=np.array(X)
    Y=np.array(Y)
    return sum(X*Y)/ len(X) - X.mean()*Y.mean()
```

```python
def col(X, Y):
    X=np.array(X)
    Y=np.array(Y)
    return cov(X,Y) / (np.std(X)*np.std(Y))
```

```python
def ajLinéaireA(x,y):
    X=np.array(x)
    Y=np.array(y)
    return cov(X,Y) / np.std(X)**2
```

```python
def ajLinéaireB(x,y):
    X=np.array(x)
    Y=np.array(y)
    return Y.mean() - X.mean()*ajLinéaireA(X,Y)
```

```python
print(cov(x,y))
print(col(x,y))
print(ajLinéaireA(x,y))
print(ajLinéaireB(x,y))
```
